```python
#Import libraries
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
from pandas import *
```

```python
#Loading data
df = pd.read_csv('../data/rfm.csv', encoding = 'utf-8',index_col=None, sep = '\t')
df.head()
```

```python
df.info()
```

```python
X = df.iloc[:, 4:7]
```

```python
#Train, val and test split
from sklearn.model_selection import train_test_split

X_train, X_val = train_test_split(X, test_size=0.3)

X_val, X_test = train_test_split(X_val, test_size=0.3)
```

```python
X_train = pd.DataFrame(X_train)
X_train
```

```python
#Standardization of our datas
from sklearn.preprocessing import StandardScaler
import pickle

# Create scaler object
scaler = StandardScaler() 
#Fit scalar object to our train set
scaler.fit(X_train)

#Save the model
#filename = 'standardization.pkl'
#pickle.dump(scaler, open(filename, 'wb'))

#Transform train, val and test into standardize datas
X_train = pd.DataFrame(scaler.transform(X_train))
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)
```

```python
from sklearn import cluster, metrics
from sklearn.cluster import KMeans

wcss = []
for i in range(1,15):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
    
silhouettes = []
for i in range(2,15):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(X_train)
    silh = metrics.silhouette_score(X_train, kmeans.labels_)
    silhouettes.append(silh)
```

```python
plt.plot(range(2,15), silhouettes)
plt.title('Silhouettes')
plt.xlabel('Cluster number')
plt.ylabel('Silhouette score')
plt.show()
```

```python
plt.plot(range(1,15), wcss)
plt.title('Elbow graph')
plt.xlabel('Cluster number')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0)

clusters = kmeans.fit_predict(X)

df['clusters'] = clusters
df.sample(10)
```
