```python
#A function which group the prices into intervals
def Group_Price(x):
    if x <= 2:
        return 2
    elif x <= 5:
        return 5
    elif x <= 10:
        return 10
    elif x <= 25:
        return 25
    elif x <= 50:
        return 50
    elif x <= 100:
        return 100
    elif x <= 300:
        return 300
```

```python
#New column group price where price are grouped in intervals
df1['group_price'] = df1['UnitPrice'].apply(Group_Price)

#Feature which count the number of item in each interval of prices
price = df1['group_price'].value_counts()
```

# Feature engineering 

## Time

```python
import datetime as dt

#Transform the column invoice date into date time object
df1['InvoiceDate'] = pd.to_datetime(df1['InvoiceDate'])
```

```python
df1.head()
```

```python
#Creation of the feature 'Month'
df1['Month'] = [x.month for x in df1.InvoiceDate]

#Dataframe of the monthly sum of prices and quantity
MonthTable = df1.groupby('Month').agg({'Quantity': lambda x: x.sum(),'UnitPrice': lambda x: x.sum()})
MonthTable
```

```python
#Monthly spending and quantity

fig, ax = plt.subplots(1,2,figsize=(10,10))

plt.subplot(211)
sns.barplot(x = MonthTable.index , y = MonthTable.UnitPrice)

plt.subplot(212)
sns.barplot(x = MonthTable.index , y = MonthTable.Quantity)


plt.xlabel('Month')


plt.show()
```

```python
#Creation of the feature 'Hour'
df1['Hour'] = [x.hour for x in df1.InvoiceDate]

#Dataframe of the hourly sum of prices and quantity
HourTable = df1.groupby('Hour').agg({'Quantity': lambda x: x.sum(), 'UnitPrice': lambda x: x.sum()})
HourTable
```

```python
#Hourly spending and quantity

fig, ax = plt.subplots(1,2,figsize=(10,10))

plt.subplot(211)
sns.barplot(x = HourTable.index , y = HourTable.UnitPrice)

plt.subplot(212)
sns.barplot(x = HourTable.index , y = HourTable.Quantity)


plt.xlabel('Hour')


plt.show()
```

```python
#Function which group the hour by intervals of 3hours from 6 a.m. to 8 p.m.
def Group_Hour(x):
    if 6 <= x <= 8:
        return 6
    elif 9<= x <= 11:
        return 9
    elif 12 <= x <= 14:
        return 12
    elif 15 <= x <= 16:
        return 15
    elif 17 <= x <= 20:
        return 17
```

```python
#Creation of the feature group_price and supression of the feature hour
df1['Hour'] = df1['Hour'].apply(Group_Hour)

df1.head()
```

### RFM segmentation

```python
#New feature total price
df1['TotalPrice'] = df1['Quantity'] * df1['UnitPrice']
```

```python
#The last date in our dataset
NOW = dt.datetime(2011,12,10)
```

```python
#Creation of the rfm table 
rfmTable = df1.groupby('CustomerID').agg({'InvoiceDate': lambda x: (NOW - x.max()).days, 'InvoiceNo': lambda x: len(x), 
                                         'TotalPrice': lambda x: x.sum()})

#Transforming float into integers
rfmTable['InvoiceDate'] = rfmTable['InvoiceDate'].astype(int)

#Renaming the columns
rfmTable.rename(columns={'InvoiceDate': 'recency', 
                         'InvoiceNo': 'frequency', 
                         'TotalPrice': 'monetary_value'}, inplace=True)
```

```python
rfmTable.head()
```

```python
#Quantiles of the rdm table
quantiles = rfmTable.quantile(q=[0.25,0.5,0.75])
quantiles = quantiles.to_dict()
```

```python
#This function returns a score between 1 and 4 according to the quantile the value belongs to.
#This one will be use for the recency since the best score is when the value is in the first quantile.
def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4

#This function returns a score between 1 and 4 according to the quantile the value belongs to.
#This one will be use for the frequency and the monetary value since the best score is when the value 
#is in the last quantile.
def FMScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
#Creation of the rfm score table using the previous funtions
rfmTable['r'] = rfmTable['recency'].apply(RScore, args=('recency',quantiles,))
rfmTable['f'] = rfmTable['frequency'].apply(FMScore, args=('frequency',quantiles,))
rfmTable['m'] = rfmTable['monetary_value'].apply(FMScore, args=('monetary_value',quantiles,))
rfmTable.head()
```

```python
rfmTable['rfm'] = rfmTable.r.map(str) + rfmTable.f.map(str) + rfmTable.m.map(str)
rfmTable['rfm_score'] = [round(x,2) for x in rfmTable[['r', 'f','m']].mean(axis=1)]

rfmTable.sample(10)
```

```python
# plot the distribution of customers over R and F
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

for i, p in enumerate(['r', 'f']) :
    parameters = {'r':'Recency', 'f':'Frequency'}
    y = rfmTable[p].value_counts().sort_index()
    x = y.index
    ax = axes[i]
    bars = ax.bar(x, y, color='silver')
    ax.set_frame_on(False)
    ax.tick_params(left=False, labelleft=False, bottom=False)
    ax.set_title('Distribution of {}'.format(parameters[p]),
                fontsize=14)
    for bar in bars:
        value = bar.get_height()
        if value == y.max():
            bar.set_color('mediumaquamarine')
        ax.text(bar.get_x() + bar.get_width() / 2,
                value - 4,
                '{}\n({}%)'.format(int(value), int(value * 100 / y.sum())),
               ha='center',
               va='top',
               color='w')
        
plt.show()
```

```python
# plot the distribution of M for RF score
fig, axes = plt.subplots(nrows=4, ncols=4,
                         sharex=False, sharey=True,
                         figsize=(10, 10))

r_range = range(1, 5)
f_range = range(1, 5)
for r in r_range:
    for f in f_range:
        y = rfmTable[(rfmTable['r'] == r) & (rfmTable['f'] == f)]['m'].value_counts().sort_index()
        x = y.index
        ax = axes[r - 1, f - 1]
        bars = ax.bar(x, y, color='silver')
        if r == 5:
            if f == 3:
                ax.set_xlabel('{}\nF'.format(f), va='top')
            else:
                ax.set_xlabel('{}\n'.format(f), va='top')
        if f == 1:
            if r == 3:
                ax.set_ylabel('R\n{}'.format(r))
            else:
                ax.set_ylabel(r)
        ax.set_frame_on(False)
        ax.tick_params(left=False, labelleft=False, bottom=False)
        ax.set_xticks(x)
        ax.set_xticklabels(x, fontsize=8)

        for bar in bars:
            value = bar.get_height()
            if value == y.max():
                bar.set_color('mediumaquamarine')
            ax.text(bar.get_x() + bar.get_width() / 2,
                    value,
                    int(value),
                    ha='center',
                    va='bottom',
                    color='k')
fig.suptitle('Distribution of M for each F and R',
             fontsize=14)
plt.tight_layout()
plt.show()
```
